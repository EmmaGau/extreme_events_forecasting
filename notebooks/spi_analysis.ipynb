```python
import xarray as xr
```

```python
spi = xr.open_dataset('/home/egauillard/data/SPI_14_cal_1940_1990_era5_MED_1degr_19400101_20240229_new.nc')
```

```python
day_of_year = spi.groupby('time.dayofyear')
```
